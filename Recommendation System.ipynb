```python
book = book.drop_duplicates().reset_index(drop=True)
```

```python
books = book.pivot_table(index='ID', columns='Book', values='Rating').reset_index(drop=True)
```

```python
# To see all columns
#pd.set_option('display.max.columns',None)
books
```

```python
books.index = book.ID.unique()
```

```python
books
```

```python
books.fillna(0, inplace=True)
```

```python
books[:5]
```

```python
#Calculating cosine similarity between users
from sklearn.metrics import pairwise_distances
from scipy.spatial.distance import cosine,correlation
```

```python
user_sim = 1-pairwise_distances(books.values,metric='cosine')
```

```python
user_sim
```

```python
user_sim_df = pd.DataFrame(user_sim)
```

```python
user_sim_df
```

```python
#Set the user and column names to userid
user_sim_df.index = book.ID.unique()
user_sim_df.columns = book.ID.unique()
```
